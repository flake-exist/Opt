```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.python.keras.utils.data_utils import Sequence
import autokeras as ak
import glob
from random import shuffle
import pandas as pd
```

```python
npy_file_list = [np.load(i,allow_pickle=True) for i in glob.glob("/home/eva_01/Documents/mi/train_data_conversion/*.npy")]
shuffle(npy_file_list)
```

```python
data = np.asarray(npy_file_list)
train_size = int(data.shape[0]*0.8)
```

```python
train_data = data[:train_size]
test_data  = data[train_size:]
```

```python
x_train = np.asarray([i[0] for i in train_data])
y_train = np.asarray([i[1] for i in train_data])
y_train_1 = y_train.reshape((y_train.shape[0],-1))

x_test = np.asarray([i[0] for i in test_data])
y_test = np.asarray([i[1] for i in test_data])
y_train_1 = y_test.reshape((y_test.shape[0],-1))
```

```python
shareReg =  ak.StructuredDataRegressor(
    loss         = "mean_squared_error",
    metrics      = "mae",
    project_name = "AtomWaffe_conversion",
    max_trials   = 200,
    directory    = "/home/eva_01/Documents/mi",
    objective    = "val_loss",
    overwrite    = False,
)
```

```python
shareReg.fit(
    x_train,
    y_train,
    epochs = 100,
    validation_split = 0.15
)
```

```python
predicted_y = shareReg.predict(x_test)
```

```python
data_answer = pd.DataFrame({'predict':[i[0] for i in predicted_y],'fact':list(y_test)})
```

```python
data_answer['diff']      = data_answer.apply(lambda row:np.absolute(row['predict'] - row['fact']),axis=1)
data_answer['err_share'] = data_answer.apply(lambda row:np.round(row['diff']/row['fact']*100,2),axis=1)
```
